```python
import scipy.stats as stats

def congruential_lineal(semilla, a, c, m, n):
    numeros = []
    x = semilla  # Valor inicial

    # Generar los n números aleatorios y normalizar entre 0 y 1
    for _ in range(n):
        x = (a * x + c) % m
        numeros.append(x / m)  # Normalizar en [0, 1]
    return numeros

# Parámetros del generador congruencial lineal
n = 100
semilla = 6438
a = 7651
c = 2908
m = 2**32

# Generar los números aleatorios
random_num = congruential_lineal(semilla, a, c, m, n)

# Aplicar la prueba de Kolmogorov-Smirnov
ks_statistic, p_value = stats.kstest(random_num, 'uniform')

# Mostrar resultados de la prueba
print(f"Estadístico de Kolmogorov-Smirnov: {ks_statistic:.6f}")
print(f"Valor p: {p_value:.6f}")

# Tomar decisión basada en el valor p
afa = 0.05  # Nivel de significancia
if p_value > afa:
    print("\nDecisión: Se acepta H0 (Los números siguen una distribución uniforme)")
else:
    print("\nDecisión: Se rechaza H0 (Los números no siguen una distribución uniforme)")

# Imprimir los números aleatorios generados
print("\nNúmeros aleatorios generados:")
for i, numero in enumerate(random_num, 1):
    print(f"Número Aleatorio {i}: {numero:.6f}")
```
