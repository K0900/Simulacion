###Contexto:
En una tienda departamental de electrónica, los tiempos de
espera de los clientes para recibir atención en el mostrador de servicio
técnico tienen una distribución lognormal. Se sabe que el tiempo
promedio de espera es de 10 minutos y una desviación estándar de 2
minutos, lo que genera una cola considerable en ciertos momentos del
día.
###Objetivo:
Generar tiempos de espera para 50 clientes en el mostrador de
servicio y analizar la duración de la fila de espera.

```python
import numpy as np
import matplotlib.pyplot as plt

#Parámetros del problema
media_espera = 10  #Tiempo promedio de espera en minutos
desviacion_espera = 2  #Desviación estándar en minutos
num_clientes = 50  #Número de clientes a simular

#Conversión de media y desviación estándar a parámetros lognormales
mu = np.log(media_espera**2 / np.sqrt(desviacion_espera**2 + media_espera**2))
sigma = np.sqrt(np.log(1 + (desviacion_espera**2 / media_espera**2)))

#Simulación de tiempos de espera para 50 clientes
tiempos_espera = np.random.lognormal(mu, sigma, num_clientes)

#Mostrar resultados
print(f"Tiempos de espera simulados para {num_clientes} clientes (primeros 10):")
print(tiempos_espera[:10])

#Grafica
plt.hist(tiempos_espera, bins=20, edgecolor='black', density=True)
plt.title("Distribución de tiempos de espera en servicio técnico (Lognormal)")
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Densidad")
plt.show()
```

Los tiempos de espera en entornos como mostradores de servicio técnico tienden a ser positivos y rara vez son exactamente iguales. La distribución lognormal es ideal para modelar variables positivas que tienen una cola larga hacia la derecha (algunos clientes pueden esperar mucho más tiempo que otros) y que se centran en un valor promedio, pero con cierta variabilidad.
