###Contexto:
Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.
###Objetivo:
Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas).

```python
import numpy as np
import matplotlib.pyplot as plt

#Parámetros del problema
tasa_llegada = 1 / 15  #1 llegada cada 15 minutos (tasa por minuto)
duracion_dia = 12 * 60  #12 horas en minutos

#Simulación de tiempos entre llegadas
tiempos_entre_llegadas = np.random.exponential(1 / tasa_llegada, size=duracion_dia // 15)

#Sumar los tiempos para obtener los momentos de llegada
momentos_llegada = np.cumsum(tiempos_entre_llegadas)
momentos_llegada = momentos_llegada[momentos_llegada <= duracion_dia]  #Solo llegadas dentro de las 12 horas

#Mostrar resultados
print(f"Total de vehículos que llegaron en 12 horas: {len(momentos_llegada)}")
print("Tiempos entre llegadas (en minutos):")
print(tiempos_entre_llegadas[:10])  #Primeros 10 tiempos

#Grafica
plt.hist(tiempos_entre_llegadas, bins=20, edgecolor='black', density=True)
plt.title("Distribución de tiempos entre llegadas de vehículos eléctricos")
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Densidad")
plt.show()
```

La distribución exponencial modela el tiempo entre eventos que ocurren de manera aleatoria y continua, en este ejercicio la tasa promedio es 1 llegada
cada 15 minutos, los tiempos entre llegadas tienen una alta probabilidad de ser cercanos a los 15 minutos, pero pueden ser menores o mayores ocasionalmente.
