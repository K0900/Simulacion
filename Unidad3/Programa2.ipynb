```python
import numpy as np
import matplotlib.pyplot as plt

lambda_inactividad = 1 / 10

def generar_tiempo_inactividad(lambda_inactividad):
    u = np.random.uniform(0, 1)
    tiempo_inactividad = -1 / lambda_inactividad * np.log(1 - u)
    return tiempo_inactividad

tiempos_inactividad = [generar_tiempo_inactividad(lambda_inactividad)
                       for _ in range(100)]

# Mostrar los tiempos de servicio
print("Tiempos de inactividad generados:", tiempos_inactividad)
plt.hist(tiempos_inactividad, bins=10, edgecolor='black', alpha=0.7)
# Graficar los tiempos de servicio simulados
plt.title('Distribución de Tiempos de Inactividad Simulados (Exponencial)')
plt.xlabel('Tiempo de inactividad (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

La distribución exponencial es importante para modelar tiempos de inactividad porque describe eventos aleatorios como fallas en equipos o tiempos entre llegadas de una manera simple y práctica, en este tipo de distribucion el pasado no influye en la probabilidad futura lo que la hace ideal para predecir fallas o mantenimientos sin complicaciones, además al depender de un solo parámetro facilita su aplicación en la vida real y sirve como base para modelos más avanzados como los procesos de Poisson.
